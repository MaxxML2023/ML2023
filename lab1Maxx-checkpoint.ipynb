```python

import numpy as np

# Create an array and a random array
arr = np.arange(start=2, stop=12, step=2)
random_arr = np.random.randint(33, size=5)
print(arr)
print(random_arr)

# Create scalar, vector, and matrix
scalar = np.array(13.0)
vector = np.arange(start=2, stop=12, step=2)
matrix = np.random.randint(low=0, high=33, size=(3, 3))

# Read and display data
print("\n* Scalar data: ", scalar)
print("* Vector data: ", vector)
print("  Vector data: ", vector[-3:-1])
print("* Matrix: ", "\n", matrix)
print("  Matrix data: ", matrix[1, 0:2])

# Update data
scalar_value = matrix[0, 0]
vector[-1] = 100
matrix[2, 2] = 369
print("\n* Updated Scalar data: ", scalar_value)
print("* Updated Vector data: ", vector)
print("* Updated Matrix: ", "\n", matrix)

# Delete data
vector = np.delete(vector, [-1])
matrix = np.delete(matrix, 1, 0)  # Delete the 2nd row
matrix = np.delete(matrix, 0, 1)  # Delete the 1st column

print("\n* Vector after delete func: ", vector)
print("* Matrix after delete func: ", "\n", matrix)


```

```python
import pandas as pd

border = '\n----------------------------------\n'

# Define your themes as separate lists
theme_1 = ["string1", "string2", "string3", "string4", "string5"]
theme_2 = [["item1", "category1"], ["item2", "category2"], ["item3", "category3"]]
theme_3 = {'Column1': ['item1', 'item2', 'item3'], 'Column2': ['category1', 'category2', 'category3']}
theme_4 = {'Column1': ['item1', 'item2', 'item3'], 'Column2': ['category1', 'category2', 'category3']}
theme_5 = [{'Column1': 'item1', 'Column2': 'category1'}, {'Column1': 'item2', 'Column2': 'category2'}, {'Column1': 'item3', 'Column2': 'category3'}]

print("1 - creating df from list\n")
df_1 = pd.DataFrame(theme_1, columns=['Column_Name_1'])
print(df_1)

print(border)
print("2 - creating df from lists of lists\n")
df_2 = pd.DataFrame(theme_2, columns=['Column_Name_2', 'Column_Name_3'])
print(df_2)

print(border)
print("3 - creating df from dict of lists\n")
df_3 = pd.DataFrame(theme_3)
print(df_3)

print(border)
print("4 - creating df by providing indexes\n")
df_4 = pd.DataFrame(theme_4, index=['Index1', 'Index2', 'Index3'])
print(df_4)

print(border)
print("5 - creating df from list of dicts\n")
df_5 = pd.DataFrame(theme_5, index=['Index_A', 'Index_B', 'Index_C'])
print(df_5)

print(border)
print("* First 3 rows of 1 df:\n")
print(df_1.head(3))

print("\n* Last 3 rows of 5 df:\n")
print(df_5.tail(3))

print("\n* Row with 'Index2' index of 4 df:\n")
print(df_4.loc['Index2'])

print("\n* Rows with 1st and 3rd indexes of 2 df:\n")
print(df_2.iloc[[0, 2]])

print("\n* Descriptive statistic about the data of 3 df")
print(df_3.describe())

df_3['New_Column'] = ['Value1', 'Value2', 'Value3']
print("\n* Changed 3 df")
print(df_3)

df_3 = df_3.drop('New_Column', axis=1)
print("\n* Back to original 3 df")
print(df_3)

print('\n* Saved data from 5 df to csv file')
df_5.to_csv('df_5.csv', index=False)
print('Load the CSV into a df:')
df_csv = pd.read_csv('df_5.csv')
print(df_csv.to_string())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define parameters
k = 3
l = 9

# Generate random data
np.random.seed(0)
x = np.linspace(0, 13, 100)
y = 3 * x + l

# Introduce measurement errors
meserr = np.random.normal(0, 3, 100)
y_meserr = y + meserr

# Create a plot
plt.figure(figsize=(10, 6))
plt.title('Linear function and data with errors')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(x, y, label='Real plot', c='indigo')
plt.scatter(x, y_meserr, label='Plot with measurement error', c='deeppink', alpha=0.5)
plt.legend()

# Calculate Mean Absolute Error (MAE) and Mean Squared Error (MSE)
mae = np.mean(np.abs(y - y_meserr))
mse = np.mean((y - y_meserr) ** 2)
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)

# Create a dataframe for results
res = {'X': x,
       'Y': y,
       'Y_hat': y_meserr,
       'MAE': mae,
       'MSE': mse}
df_res = pd.DataFrame(res)

# Save results to a CSV file
df_res.to_csv('df_res.csv', index=False)

# Show the plot
plt.show()
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Add Pandas import

k = 3
l = 9

np.random.seed(0)
x = np.linspace(0, 13, 100)
y = 3 * x + l

meserr = np.random.normal(0, 3, 100)
y_meserr = y + meserr

plt.figure(figsize=(10, 6))
plt.title('Linear function and data with errors')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(x, y, label='Real plot', c ='indigo')
plt.scatter(x, y_meserr, label='Plot with measurement error', c='deeppink', alpha=0.5)
plt.legend()

mae = np.mean(np.abs(y - y_meserr))
mse = np.mean((y - y_meserr) ** 2)
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)

res = {'X': x,
       'Y': y,
       'Y_hat': y_meserr,
       'MAE': mae,  # Changed 'mAE' to 'MAE'
       'MSE': mse}  # Changed 'mSE' to 'MSE'
df_res = pd.DataFrame(res)
df_res.to_csv('df_res.csv', index=False)

plt.show()
import numpy as np
import matplotlib.pyplot as plt

k = 3
l = 9

np.random.seed(0)
x = np.linspace(0, 13, 100)
y = 3 * x + l

meserr = np.random.normal(0, 3, 100)
y_meserr = y + meserr

plt.figure(figsize=(10, 6))
plt.title('Linear function and data with errors')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.plot(x, y, label='Real plot', c ='indigo')
plt.scatter(x, y_meserr, label='Plot with measurement error', c='deeppink', alpha=0.5)
plt.legend()

mae = np.mean(np.abs(y - y_meserr))
mse = np.mean((y - y_meserr) ** 2)
print("Mean absolute error (mae):", mae)
print("Mean squared error (mse):", mse)

res = {'X': x,
            'Y': y,
            'Y_hat': y_meserr,
            'mAE': mae,
            'mSE': mse}
df_res = pd.DataFrame(res)
df_res.to_csv('df_res.csv', index=False)

plt.show()
```
