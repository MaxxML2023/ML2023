```python
import numpy as np
import random
import matplotlib.pyplot as plt

# Створюємо синтетичкий набір даних
np.random.seed(0)
X = np.linspace(0, 1, 100)  # Вхідні ознаки (приклад: час)
y = 2 * X + 1 + np.random.normal(0, 0.2, 100)  # Вихідні значення (приклад: ціна)

# Розділяємо дані на навчальний та тестовий набори
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# Візуалізуємо дані
plt.scatter(X_train, y_train, label='Навчальні дані')
plt.scatter(X_test, y_test, label='Тестові дані')
plt.xlabel('Вхідні ознаки')
plt.ylabel('Вихідні значення')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Побудова моделі нейронної мережі
model = keras.Sequential()
model.add(layers.Dense(4, activation='relu', input_shape=(1,)))
model.add(layers.Dense(4, activation='relu'))
model.add(layers.Dense(1))

# Скомпілюємо модель
model.compile(optimizer='adam', loss='mean_squared_error')

# Тренування моделі
history = model.fit(X_train, y_train, epochs=200, verbose=0, validation_data=(X_test, y_test))

# Візуалізація зміни функції втрат на тренувальному та тестовому наборах
plt.plot(history.history['loss'], label='Функція втрат на навчальних даних')
plt.plot(history.history['val_loss'], label='Функція втрат на тестових даних')
plt.xlabel('Епохи')
plt.ylabel('Функція втрат')
plt.legend()
plt.show()

# Оцінка моделі
loss = model.evaluate(X_test, y_test, verbose=0)
print(f'Загальна функція втрат на тестових даних: {loss:.4f}')
```
