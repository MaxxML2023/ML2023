```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Завантаження даних для бінарної класифікації
binary_data = pd.read_csv('binary_classification_data.csv')

# Розділення на атрибути та цільову змінну
X_binary = binary_data[['Attribute1', 'Attribute2']]
y_binary = binary_data['Label']

# Розділення на навчальний та тестувальний набори
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(X_binary, y_binary, test_size=0.2, random_state=42)

# Побудова моделі для бінарної класифікації
binary_model = Sequential([
    Dense(64, activation='relu', input_shape=(2,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Компіляція моделі для бінарної класифікації
binary_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Навчання моделі для бінарної класифікації
history_binary = binary_model.fit(X_train_binary, y_train_binary, epochs=50, batch_size=32, validation_data=(X_test_binary, y_test_binary), verbose=0)

# Візуалізація функції втрат та точності для бінарної класифікації
plt.figure(figsize=(12, 4))

# Графік функції втрат
plt.subplot(1, 2, 1)
plt.plot(history_binary.history['loss'], label='Train')
plt.plot(history_binary.history['val_loss'], label='Test')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Binary Classification - Loss')
plt.legend()

# Графік точності
plt.subplot(1, 2, 2)
plt.plot(history_binary.history['accuracy'], label='Train')
plt.plot(history_binary.history['val_accuracy'], label='Test')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Binary Classification - Accuracy')
plt.legend()

plt.tight_layout()
plt.show()

```
